```python
import pandas as pd                # Pandas를 호출하여 데이터 프레임 핸들링
import seaborn as sns              # Seaborn을 호출하여 다양한 통계적 그래프 사용
import matplotlib.pyplot as plt    # matplot을 호출하여 기본 그래프 준비

# 머신러닝 패키지 호출

from sklearn.linear_model import LogisticRegression  # 로지스틱 회귀 호출
from sklearn import tree                             # 의사결정나무(Decision tree) 호출
from sklearn.ensemble import RandomForestClassifier  # Random Forest 호출
from sklearn import svm                              # 서포트 벡터 호출
from sklearn import neural_network                   # 신경망 호출
```

```python
# 구글 드라이브를 연동하여 Data 읽어오기

Train_IRIS = '/content/drive/My Drive/Colab Notebooks/Train_IRIS.csv'
Test_IRIS = '/content/drive/My Drive/Colab Notebooks/Test_IRIS.csv'
Full_IRIS = '/content/drive/My Drive/Colab Notebooks/Full_IRIS.csv'

df_Train = pd.read_csv(Train_IRIS)
df_Test = pd.read_csv(Test_IRIS)
df_Full = pd.read_csv(Full_IRIS)

# 상관관계 그려보기
sns.pairplot(data = df_Full, hue='Species')   # target 컬럼의 이름이 Species
plt.show()
```

```python
# 훈련데이터(Training Data)와 Test data 분리하기
X_Train = df_Train[['Petal.Length','Petal.Width','Sepal.Length','Sepal.Width']]
y_Train = df_Train['Species']
X_Test = df_Test[['Petal.Length','Petal.Width','Sepal.Length','Sepal.Width']]
y_Test = df_Test['Species']
```

```python
# 다항 로지스틱 회귀
ML_m = LogisticRegression().fit(X_Train,y_Train)
ML_p = ML_m.predict(X_Test)
print("다항 로지스틱 회귀")
print(pd.crosstab(ML_p,y_Test))
```

```python
# Decision Tree
tree_m = tree.DecisionTreeClassifier().fit(X_Train,y_Train)
tree_p = tree_m.predict(X_Test)
print("의사 결정 나무")
print(pd.crosstab(tree_p,y_Test))
```

```python
# Random Forest
rf_m = RandomForestClassifier().fit(X_Train,y_Train)
rf_p = rf_m.predict(X_Test)
print("Random Forest")
print(pd.crosstab(rf_p,y_Test))
```

```python
# Support Vector
sv_m = svm.SVC().fit(X_Train,y_Train)
sv_p = sv_m.predict(X_Test)
print("Support Vector")
print(pd.crosstab(sv_p,y_Test))
```

```python
# Neural network
Nn_m = neural_network.MLPClassifier(max_iter=5000).fit(X_Train,y_Train)
Nm_p = Nn_m.predict(X_Test)
print("Neural network")
print(pd.crosstab(Nm_p,y_Test))
```

```python
# Decision Tree
tree_d = tree.DecisionTreeClassifier().fit(X_Train,y_Train)
tree_pre = tree_d.predict(X_Test)
tree.plot_tree(tree_d)

print("Decision Tree")
print(pd.crosstab(tree_pre,y_Test))
plt.show()
```
